# HIERARCHICAL CLUSTERING 


```python
# Normalization 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
airline1=(airline.iloc[:,1:])
airline1.head()
```

```python
# Normalized data frame (considering the numerical part of data)
airline1_norm = norm_func(airline1.iloc[:,1:])
```

```python
airline1_norm
```

```python
plt.figure(figsize=(15,12));plt.title('HIERARCHY DENDROGRAM');plt.xlabel('Index');plt.ylabel('Distance')
dendograms=sch.dendrogram(sch.linkage(airline1_norm,"complete"))
plt.show()
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'single')
```

```python
hc
```

```python
# save clusters for chart
y_hc = hc.fit_predict(airline1_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
import pandas as pd
import matplotlib.pylab as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist 
import numpy as np
```

```python
airl=pd.read_excel(R'D:\DATA SCIENCE ASSIGNMENTS\7. CLUSTERING\EastWestAirlines.xlsx',sheet_name='data')#assigning it to airl so that the machine doesnt confuse with the above airline
```

```python
# Normalization 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)

# Normalized data frame 
df_norm = norm_func(airl.iloc[:,1:])


df_norm.head(10)  # Top 10 rows
```

```python
# Selecting 5 clusters from the above scree plot which is the optimum number of clusters 
model=KMeans(n_clusters=5) 
model.fit(df_norm)

model.labels_ # getting the labels of clusters assigned to each row
```

```python
md=pd.Series(model.labels_)  # converting numpy array into pandas series object 
airl['clust']=md # creating a  new column and assigning it to new column 
df_norm.head()
```

```python
airl.iloc[:,1:7].groupby(airl.clust).mean()
```

```python
airl.head()
```

```python
Wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    Wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),Wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('Wcss')
plt.show()
```

# DBSCAN

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import .csv file and convert it to a DataFrame object
airl=pd.read_excel(r'D:\DATA SCIENCE ASSIGNMENTS\7. CLUSTERING\EastWestAirlines.xlsx',sheet_name='data')

print(airl.head())
```

```python
print(airl.info())
```

```python
airl.drop(['ID#','Award?'],axis=1,inplace=True)
```

```python
airl
```

```python
array=airl.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=1, min_samples=12)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([airl,cl],axis=1)
```
